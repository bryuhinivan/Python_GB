**Задача №67.** Решение в группах
1. Создать новый столбец в таблице с
пингвинами, который будет отвечать за
показатель длины клюва пингвина.
high - длинный(от 42)
middle - средний(от 35 до 42)
low - маленький(до 35)

```python
penguins.columns
```

```python
# Создать новый столбец в таблице с пингвинами, который будет отвечать за показатель длины клюва пингвина.
# high - длинный(от 42)
# middle - средний(от 35 до 42)
# low - маленький(до 35)
penguins.loc[penguins['bill_length_mm'] > 42, 'group'] = 'high'
penguins.loc[(penguins['bill_length_mm'] >= 35) & (penguins['bill_length_mm'] <= 42), 'group'] = 'middle'
penguins.loc[penguins['bill_length_mm'] < 35, 'group'] = 'low'
```
